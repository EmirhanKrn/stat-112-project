```
data.columns=data.columns.str.title()
data.columns
```

```
data=data.rename({"Marital_Status":"Marital Status"}, axis="columns")
data=data.rename({"Perform_Evaluation":"Perform Evaluation"}, axis="columns")
data=data.rename({"Monthly_Income":"Monthly Income"}, axis="columns")
data=data.rename({"Working_Years":"Working Years"}, axis="columns")
data=data.rename({"Education_Level":"Education Level"}, axis="columns")
data.columns
```

```
for i in data.select_dtypes(include="object").columns:
  print(data[i].value_counts())
```

```
data["Gender"].loc[data["Gender"]=="female"]="Female"
data["Gender"].loc[data["Gender"]=="male"]="Male"
data["Gender"].loc[data["Gender"]=="MaLe"]="Male"
data["Gender"].loc[data["Gender"]=="feMaLe"]="Female"
data["Gender"].value_counts()
```

```
data["Race"].loc[data["Race"]=="WHITE"]="White"
data["Race"].loc[data["Race"]=="Black or AFRican American"]="Black or African American"
data["Race"].loc[data["Race"]=="ASİan"]="Asian"
data["Race"].loc[data["Race"]=="Whİte"]="White"
data["Race"].loc[data["Race"]=="Black or Afrİcan Amerİcan"]="Black or African American"
data["Race"].value_counts()
```

```
data["Marital Status"].loc[data["Marital Status"]=="yes"]="Yes"
data["Marital Status"].loc[data["Marital Status"]=="no"]="No"
data["Marital Status"].value_counts()

```

```
data["Position"].loc[data["Position"]=="Production Technician DEPPartment"]="Production Technician Department"
data["Position"].loc[data["Position"]=="IT DEPPartment"]="IT Department"
data["Position"].loc[data["Position"]=="Sales DEPPartment"]="Sales Department"
data["Position"].value_counts()
```

```
data["Remote"].loc[data["Remote"]=="yes"]="Yes"
data["Remote"].loc[data["Remote"]=="no"]="No"
data["Remote"].value_counts()
```

```
for i in data.select_dtypes(include="object").columns:
  print(data[i].value_counts())
```

```
data["Gender"]=data["Gender"].fillna("Female")
data["Race"]=data["Race"].fillna("Black or African American")
data["Marital Status"]=data["Marital Status"].fillna("Yes")
data["Position"]=data["Position"].fillna("Production Technician Department")
data["Education Level"]=data["Education Level"].fillna("University")
data["Location"]=data["Location"].fillna("West South Central")
data["Remote"]=data["Remote"].fillna("Yes")
```

```
data
```

```
data.describe().transpose()
```

```
data["Perform Evaluation"].loc[data["Perform Evaluation"]==data["Perform Evaluation"].max()]=0
mean_Pe = data['Perform Evaluation'].mean()
mean_Pe1=mt.ceil(mean_Pe)
data["Perform Evaluation"].loc[data["Perform Evaluation"]==0]=mean_Pe1
data['Perform Evaluation'] = data['Perform Evaluation'].fillna(mean_Pe1)
data["Perform Evaluation"].head(20)
```

```
data["Monthly Income"].loc[data["Monthly Income"]==data["Monthly Income"].max()]=0
mean_inc=data["Monthly Income"].mean()
data["Monthly Income"].loc[data["Monthly Income"]==0]=mean_inc
data["Monthly Income"]=data["Monthly Income"].fillna(mean_inc)
data["Monthly Income"].head(14)
```

```
data["Working Years"].loc[data["Working Years"]==data["Working Years"].max()]=0
mean_working = data['Working Years'].mean()
mean_working1=mt.ceil(mean_working)
data["Working Years"].loc[data["Working Years"]==0]=mean_working1
data['Working Years'] = data['Working Years'].fillna(mean_working1)
data["Working Years"].head(18)
```

```
data["Age"].loc[data["Age"]==data["Age"].max()]=0
mean_age=data["Age"].mean()
mean_age1=mt.ceil(mean_age)
mean_working1=mt.ceil(mean_working)
data["Age"].loc[data["Age"]==0]=mean_age1
data["Age"]=data["Age"].fillna(mean_age1)
data["Age"].head(5)
```
