```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
data = pd.read_excel('datacleaned.xlsx')
```

```
#Which district take more monthly income?
#West South Central and East Norh Central have most perform evaluation value. 
#Third  New England and forth West North Central.If I choose one of the 2 people with equal benefits, I choose one  form West South Central and East North Central.
sns.barplot(data=data, x="Location", y="Perform Evaluation", color="darkgreen",ci=None)
sns.set(rc={'figure.figsize':(15,7.5)})

plt.title("Realtionship Between Perform Evaluation and Location", fontsize=14, fontweight="bold")
```

```
#How effects male performans their salerys and are they deserve to their salerys?
#Male take more salery than female. Male's salery is not enough for their performance evaluation.
# Against the female, male take less salery based on their performance.
sns.barplot(data=data, x="Gender", y="Monthly Income", color="blue",ci=None)
sns.set(rc={'figure.figsize':(15,7.5)})

plt.title("Realtionship Between Monthly Income and Gender", fontsize=14, fontweight="bold")
```

```
sns.barplot(data=data, x="Gender", y="Perform Evaluation",ci=None, palette=["#ff9d9a","#72a3ff"])
sns.set(rc={'figure.figsize':(15,7.5)})
plt.legend(loc='upper right')


plt.title("Realtionship Between Perform Evaluation and Gender", fontsize=14, fontweight="bold")
```

```
data['Perfomance Evaluation/Monthly Income']=data['Perform Evaluation']/data['Monthly Income']
sns.barplot(data=data, x="Gender", y="Perfomance Evaluation/Monthly Income", color="blue",ci=None, palette=["#ff9d9a","#72a3ff"])
sns.set(rc={'figure.figsize':(15,7.5)})

plt.title("Realtionship Between Perform Evaluation/Monthly Income and Gender", fontsize=14, fontweight="bold")
```
