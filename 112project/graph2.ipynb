```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy import stats
```

```python
  data = pd.read_excel('datacleaned.xlsx')
```

```python
data.head(10)
```

```python
data.info()
```

```python
#Is there any linear relation between monthly income and age ?
plt.figure(figsize = (8,8))
sns.lineplot(data=data, x="Age", y="Monthly Income")
plt.title('The Relationship between Ages and Monthly Income',fontsize = 8)
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.show()
```

```python
slope,intercept,r_value,p_value,std_err=stats.linregress(data['Perform Evaluation'],data['Monthly Income'])
```

```python
plt.figure(figsize = (8,8))
plt.xlabel('Perform Evaluation',fontsize=12,fontweight='bold')
plt.ylabel('Monthly Income',fontsize=12,fontweight='bold')


ax=sns.regplot(data=data,x='Perform Evaluation',y='Monthly Income',line_kws={'Label':'y={0:.1f}x+{1:.1f}'.format(slope,intercept)})
ax.legend()
```
